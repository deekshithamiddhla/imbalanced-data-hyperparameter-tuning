```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import random
import matplotlib.pyplot as plt

X = random.sample(range(10, 59), 25)
print(X)
Y = random.sample(range(1, 65), 25)
print(Y)

m = 0
c = 0
L = 0.0001
epochs = 100
n = float(len(X))
X = [float(x) for x in X]  # Convert X to float using list comprehension
# Initialize Y_pred
Y_pred = [m * x + c for x in X]
#Y_pred = m * X + c
m_values = []  # To store the values of m in each epoch
c_values = []  
loss_values = [] 
for i in range(epochs):
    D_m = (-2/n) * sum(X[j] * (Y[j] - Y_pred[j]) for j in range(len(X)))
    D_c = (-2/n) * sum(Y[j] - Y_pred[j] for j in range(len(X)))
    m = m - L * D_m
    c = c - L * D_c
    Y_pred = [m * x + c for x in X]
    #Y_pred = m * X + c
     # Calculate the loss function
    Y_np = np.array(Y, dtype=float)  # Convert Y to a NumPy array
    loss = np.mean((Y_np - Y_pred) ** 2)
    loss_values.append(loss)
    m_values.append(m)
    c_values.append(c)
    
    print(f"Epoch {i+1}: m = {m}, c = {c},loss={loss}")

#plt.scatter(X, Y)
#plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')
#plt.show()
```

```python
# Gradient Descent graph
plt.subplot(1, 2, 2)
plt.plot(range(1, epochs+1), m_values, label='m')
plt.plot(range(1, epochs+1), c_values, label='c')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Gradient Descent')
plt.legend()

plt.tight_layout()
plt.show()
```
